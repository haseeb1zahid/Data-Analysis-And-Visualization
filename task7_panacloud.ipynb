```python
import pandas as pd
import matplotlib.pyplot as plt


myfile = pd.read_csv("C:/Users/Admin/PycharmProjects/Panacloud_Bootcamp/coinmarketcap_06122017.csv")
volatility = myfile[['id', 'percent_change_24h', 'percent_change_7d']].set_index('id').dropna().sort_values(
    by='percent_change_24h')
# cap10 = volatility[:10]
# print(cap10)

def top_10_subplot(volatility_series, title):

    # Making the subplot and the figure for two side by side plots
    fig, ax1 = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    ax1[0].bar(volatility.index[:10], volatility[volatility_series][:10])
    ax1[0].set_title("Losers")
    ax1[0].tick_params(labelrotation=90)
    ax1[0].set_ylabel('% change')
    ax1[1].bar(volatility.index[-10:], volatility[volatility_series][-10:])
    ax1[1].set_title("Winners")
    ax1[1].tick_params(labelrotation=90)
    ax1[1].set_ylabel('% change')
    fig.suptitle(title, fontsize=14)
    plt.show()

DTITLE = "24 hours top losers and winners"
top_10_subplot('percent_change_24h', DTITLE)
```
